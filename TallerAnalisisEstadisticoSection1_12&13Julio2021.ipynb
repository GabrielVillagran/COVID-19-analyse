**Caso de Estudio** Considerar la base de datos generada por CONACYT acerca de casos confirmados, sospechosos, negativos y defunciones a nivel nacional y estatal de COVID-19: 

https://datos.covid-19.conacyt.mx/

Realizar una evaluación del número de casos confirmados y defunciones, y el porcentaje de defunciones por casos confirmados por COVID-19 en el estado de San Luis Potosí, y comparar con algunos estado vecinos:

+ *¿Con cual estado existen una mayor similitud en la evolución de casos confirmados?*
+ *¿Con cual estado existen una mayor similitud en la evolución de las defunciones?*
+ *¿Con cual estado existen una mayor similitud en el porcentaje de defunciones por casos confirmados?*

***

<font color=blue>**Paso 1**</font>: Cargar las bases de datos que se decargaron de la página web en formato CSV

```r
confirmados=read.csv('Casos_Confirmados.csv')
defunciones=read.csv('Casos_Defunciones.csv')
```

<font color=blue>**Paso 2**</font>: Visualizar las base de datos de **casos confirmados** y obtener sus dimensiones

```r
head(confirmados)
tail(confirmados)
confirmados
print('Numero de renglones')
nrow(confirmados)
print('Numero de columnas')
ncol(confirmados)
```

<font color=blue>**Paso 3**</font>: Graficar toda la información de **casos confirmados** en el estado de San Luis Potosi

```r
x=confirmados[confirmados$nombre=='SAN LUIS POTOSI',4:ncol(confirmados)]
dias=seq(1,ncol(x))
plot(dias,x,type='b',pch=18, lty=4, lwd=0.5, main='Casos Confirmados en San Luis Potosi', 
     xlab='Dias', ylab='Número de Casos Diarios', col=2)
grid()
```

<font color=blue>**Paso 4**</font>: Graficar la evolución del **acumulado de casos confirmados** en San Luis Potosí, y en algunos estados vecinos

```r
x=confirmados[confirmados$nombre=='SAN LUIS POTOSI',4:ncol(confirmados)]
dias=seq(1,ncol(x))
index=1
plot(dias,cumsum(t(x)),type='l', lwd=2, main='Casos Acumulados San Luis Potosi', 
     xlab='Dias', ylab='Número de Casos Acumulados', col=index, lty=index)

index=index+1
estados=c('NUEVO LEON', 'COAHUILA', 'TAMAULIPAS', 'VERACRUZ', 'QUERETARO')
for(state in estados){
    y=confirmados[confirmados$nombre==state,4:ncol(confirmados)]
    lines(dias,cumsum(t(y)), col=index, lty=index, lwd=2)
    index=index+1
}
legend('topleft', c('SAN LUIS POTOSI', estados), col=seq(1,index-1), lty=seq(1,index-1), lwd=2)
grid()
```

<font color=blue>**Paso 5**</font>: Evaluar el **porcentaje del acumulado de casos positivos por COVID-19 con respecto de la población total** del mismo estado

```r
p=confirmados[confirmados$nombre=='SAN LUIS POTOSI',2]
z=cumsum(t(x))/p
index=1
plot(dias,z,type='l', lwd=2, main='Casos Acumulados San Luis Potosi con respecto a la Población del estado', 
     xlab='Dias', ylab='Número de Casos Acumulados/ Población del Estado', col=index, lty=index)

index=index+1
for(state in estados){
    y=confirmados[confirmados$nombre==state,4:ncol(confirmados)]
    q=confirmados[confirmados$nombre==state,2]
    w=cumsum(t(y))/q
    lines(dias,w, lwd=2, col=index, lty=index)
    index=index+1
}
legend('topleft', c('SAN LUIS POTOSI', estados), col=seq(1,index-1), lty=seq(1,index-1), lwd=2)
grid()
```

<font color=blue>**Paso 6**</font>: Evaluar la **similitud** en la evolución de la pandemia con respecto del acumulado de los casos confirmados con respecto de la población estatal a través del **coeficiente de determinación** $\chi^2$ (*valor limitado al intervalo -1 a +1*).

```r
m=length(estados)
similitud=matrix(ncol=m)
index=1
for(state in estados){
    y=confirmados[confirmados$nombre==state,4:ncol(confirmados)]
    q=confirmados[confirmados$nombre==state,2]
    w=cumsum(t(y))/q
    similitud[index]=100*(cor(z,w)**2)
    index=index+1
}
print(estados)
print(similitud)
```

<font color=blue>**Paso 7**</font>: Visualizar las base de datos de **defunciones** por COVID-19 y obtener sus dimensiones

```r
head(defunciones)
tail(defunciones)
print('Número de renglones')
nrow(defunciones)
print('Numero de columnas')
ncol(defunciones)
```

<font color=blue>**Paso 9**</font>: Graficar toda la información de defunciones en el estado de San Luis Potosi

```r
x=defunciones[defunciones$nombre=='SAN LUIS POTOSI',4:ncol(defunciones)]
dias=seq(1,ncol(x))
plot(dias,x,type='b',pch=18, lty=4, lwd=0.5, main='Defunciones por COVID-19 en San Luis Potosi', 
     xlab='Dias', ylab='Número de Defunciones', col=2)
grid()
```

<font color=blue>**Paso 10**</font>: Graficar el acumulado de **defunciones en el estado de San Luis Potosi con respecto de la población total** del estado y comparar con otros estados, así como cuantificar su **similitud** por medio del coeficiente de determinación $\chi^2$ (*valor limitado al intervalo -1 a +1*).

```r
x=defunciones[defunciones$nombre=='SAN LUIS POTOSI',4:ncol(defunciones)]
dias=seq(1,ncol(x))
p=defunciones[defunciones$nombre=='SAN LUIS POTOSI',2]
z=cumsum(t(x))/p
index=1
plot(dias,z,type='l', lwd=1.5, main='Defunciones por COVID-19 en San Luis Potosi / Población', 
     xlab='Dias', ylab='Número de Defunciones / Población del Estado', col=index, lty=index)

similitud=matrix(ncol=m)
for(state in estados){
    y=defunciones[defunciones$nombre==state,4:ncol(defunciones)]
    q=defunciones[defunciones$nombre==state,2]
    w=cumsum(t(y))/q
    similitud[index]=100*(cor(z,w)**2)
    lines(dias,w, lwd=1.5, col=index+1, lty=index+1)
    index=index+1
}
legend('topleft', c('SAN LUIS POTOSI', estados), col=seq(1,index-1), lty=seq(1,index-1), lwd=1.5)
grid()
print(estados)
print(similitud)
```

<font color=blue>**Paso 11**</font>: Calcular el **porcentaje total de defunciones por casos acumulados** en el estado de San Luis Potosi, y comparar con los estados vecinos analizados previamente.

```r
x=confirmados[confirmados$nombre=='SAN LUIS POTOSI',4:ncol(confirmados)]
y=defunciones[defunciones$nombre=='SAN LUIS POTOSI',4:ncol(defunciones)]
porcentaje=100*sum(y)/sum(x)
print('Porcentaje de defunciones por casos confirmados en SLP =')
print(porcentaje)
similitud=matrix(ncol=m)
index=1
for(state in estados){
    y=defunciones[defunciones$nombre==state,4:ncol(defunciones)]
    x=confirmados[defunciones$nombre==state,4:ncol(defunciones)]
    w=sum(y)/sum(x)
    similitud[index]=100*w
    index=index+1
}
print(estados)
print(similitud)
```
